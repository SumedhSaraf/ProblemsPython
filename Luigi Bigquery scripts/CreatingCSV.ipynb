```python
import cv2
import numpy as np
import pandas as pd
```

```python
import os
images = os.listdir("./data5/")
for imageName in images:
    print(imageName)
    imagepath = "./data5/" + imageName
    print(imagepath)
    img_rgb = cv2.imread(imagepath)
    print(img_rgb.shape)
    small = cv2.resize(img_rgb, (0,0), fx=0.05, fy=0.05)
    temp = small
    small1 = temp.reshape(1,6912)
    c = np.append(c,small1, axis=0)
    
columns = list(range(6912))  
df2 = pd.DataFrame(columns=columns,data = c)
df2.shape
```
